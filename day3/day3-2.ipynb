```python
def get_mode(arr):
    cs = {'0': 0, '1': 0}
    for c in arr:
        if c == '0':
            cs['0'] += 1
        else:
            cs['1'] += 1
    
    if cs['0'] > cs['1']:
        return '0'
    elif cs['1'] > cs['0']:
        return '1'
    else:
        return '1'
    
    
def get_reverse_mode(arr):
    if get_mode(arr) == '1':
        return '0'
    else:
        return '1'
    
    
def get_oxygen(nums):
    index = 0
    while len(nums) > 1:
        pref_num = get_mode([x[index] for x in nums])
        print("index: {}    preffered num: {}".format(index, pref_num))
        bad_nums = [x for x in nums if x[index] != pref_num]
        nums = [x for x in nums if x not in bad_nums]
        index += 1
    return nums[0]


def get_co2(nums):
    index = 0
    while len(nums) > 1:
        pref_num = get_reverse_mode([x[index] for x in nums])
        print("index: {}    preffered num: {}".format(index, pref_num))
        bad_nums = [x for x in nums if x[index] != pref_num]
        nums = [x for x in nums if x not in bad_nums]
        index += 1
    return nums[0]
    

def main():
    with open("day3input.txt") as file:
        lines = [x.strip('\n') for x in file.readlines()]
    
    for line in lines[0:10]: print(line)
    o = get_oxygen(lines)
    c = get_co2(lines)
    l = int(o, 2) * int(c, 2)
    print('oxygen rating: {}'.format(o))
    print('co2 rating: {}'.format(c))
    print('life support rating: {}'.format(l))
    

if __name__ == "__main__":
    main()
```
